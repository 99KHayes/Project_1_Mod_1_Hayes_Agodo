## Bom Movie Set Studio Information

#### This section tried to answer the question of where should the new streaming company produce their movies. The Bom movie set has the studio and gross profit information for a little over 3,000 movies.

##### Analysis Process
1. This data set was analyzed by trying to find the unique studios and counting how many movies were produced at each studio. Only domestic profit was considered as the streaming service is targeted at the US making foregn profit irrelevant.
2. The top numerical producing studios were graphed first quick and dirty and matlab and then a more refined seaborn plot of the number of movies per studio.
3. Next the rows with null values were removed as well as the year column. It is known all the movies are on the correct 2010 - 2018 time period from the information on the download page for the data set.
4. THe data was grouped by the studio and the average dometic gross was calculated for each studio. It was then divided by 1,000,000 in order to make the graphs and data more readable to humans.

```python
# Get the average domestic budget for each studio and divide it by a million 
bom_movie_neatened = bom_movie_no_nulls.groupby('studio').mean() /1000000
```

### Lets Look at the categories of behind the scenes roles


```python
final.groupby('category').mean()
```

```python
#Sorting for top directors
director_rating = final[final.category == 'director']

director_rating = director_rating.groupby('primary_name').mean().sort_values(['averagerating','numvotes'], ascending = False)
director_rating = director_rating[(director_rating.start_year < 2019.0) & (director_rating.start_year >= 2010.0) & (director_rating.numvotes > 4110)]
top_director = director_rating.head(5)
top_director
```

```python
#Sorting for top producer
producer_rating = final[final.category == 'producer']

producer_rating = producer_rating.groupby('primary_name').mean().sort_values(['averagerating','numvotes'], ascending = False)
producer_rating = producer_rating[(producer_rating.start_year < 2019.0) & (producer_rating.start_year >= 2010.0)& (producer_rating.numvotes > 4110)]
top_producer = producer_rating.head(5)
top_producer
```

```python
#Sorting for top composer
composer_rating = final[final.category == 'composer']

composer_rating = composer_rating.groupby('primary_name').mean().sort_values(['averagerating','numvotes'], ascending = False)
composer_rating = composer_rating[(composer_rating.start_year < 2019.0) & (composer_rating.start_year >= 2010.0) & (composer_rating.numvotes > 4110)]
top_composer = composer_rating.head(5)
top_composer
```

```python
#Sorting for top writer
writer_rating = final[final.category == 'writer']

writer_rating = writer_rating.groupby('primary_name').mean().sort_values(['averagerating','numvotes'], ascending = False)
writer_rating = writer_rating[(writer_rating.start_year < 2019.0) & (writer_rating.start_year >= 2010.0) & (writer_rating.numvotes > 4110)]
top_writer = writer_rating.head(5)
top_writer
```

```python
# John Williams average rating as a composer
final[(final.primary_name == 'John Williams') & (final.category == 'composer')].averagerating.mean()
```

```python
# Plot the 4 Behind the Scenes Reccomendations

fig, axs = plt.subplots(2, 2,figsize=(30,30))
fig.suptitle('Top 5 Rated Behind The Scenes Roles \n With More Than The Average Number of Reviews \nOn a scale of 1 -10', fontsize=30)

axs[0, 0].bar(top_director.index, top_director.averagerating.values, color ='green')
axs[0, 0].set_title('Top 5 Rated Directors', fontsize=30)
axs[0, 0].set_xlabel('Director', fontsize=25)
axs[0, 0].set_ylabel('Rating ', fontsize=25)
axs[0, 0].set_xticklabels(top_director.index, rotation=15, ha="right",fontsize=15)
axs[0, 0].set_yticklabels(range(5,10,1),fontsize=15)

axs[0, 1].bar(top_producer.index, top_producer.averagerating.values, color ='green')
axs[0, 1].set_title('Top 5 Rated Producers', fontsize=30)
axs[0, 1].set_xlabel('Producer', fontsize=25)
axs[0, 1].set_ylabel('Rating', fontsize=25)
axs[0, 1].set_xticklabels(top_producer.index, rotation=15, ha="right",fontsize=15)
axs[0, 1].set_yticklabels(range(5,10,1),fontsize=15)


axs[1, 0].bar(top_composer.index, top_composer.averagerating.values, color ='green')
axs[1, 0].set_title('Top 5 Rated Composers', fontsize=30)
axs[1, 0].set_xlabel('Composer', fontsize=25)
axs[1, 0].set_ylabel('Rating ', fontsize=25)
axs[1, 0].set_xticklabels(top_composer.index, rotation=15, ha="right",fontsize=15)
axs[1, 0].set_yticklabels(range(5,10,1),fontsize=15)
                     
axs[1, 1].bar(top_writer.index, top_writer.averagerating.values, color ='green')
axs[1, 1].set_title('Top 5 Rated Writers', fontsize=30)
axs[1, 1].set_xlabel('Writer', fontsize=25)
axs[1, 1].set_ylabel('Rating ', fontsize=25)
axs[1, 1].set_xticklabels(top_writer.index ,rotation=15, ha="right",fontsize=15)
axs[1, 1].set_yticklabels(range(5,10,1),fontsize=15)

plt.savefig('Top5s.png')
plt.show()
```
