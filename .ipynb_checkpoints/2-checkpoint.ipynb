# Checking movie info and reviews

```python
rt_movie_info = pd.read_csv('data/rt.movie_info.tsv.gz', delimiter = '\t')
print(rt_movie_info.info())
rt_movie_info.head()
```

```python
rt_reviews = pd.read_csv('data/rt.reviews.tsv.gz', delimiter = '\t', encoding= 'unicode_escape')
print(rt_reviews.info())
rt_reviews.head()
```

# Reviews Exploration

```python
#check top movie site publishers
rt_reviews['publisher'].value_counts().head()
```

```python
#graph of movie site publisher data
rt_reviews['publisher'].value_counts().head().plot.bar(color='green')
plt.title('Top 5 Movie Site Reviewers', fontsize=16)
plt.xlabel('Reviewer Source', fontsize=12)
plt.ylabel('Number of Review', fontsize=12)
plt.xticks(rotation=45)
```

```python
#sort date in ascending order to see earliest date
rt_reviews['date'].sort_values(ascending=True).head()
```

```python
#slice data so only year is selected
rt_reviews['date'].map(lambda x: x[-4:]).head()
```

```python
#change date to string datatype
rt_reviews['date']= rt_reviews['date'].astype('str')
```

```python
rt_reviews['date'].map(lambda x: x[-4:]).head()
```

```python
#make new column for year only
rt_reviews['year'] = rt_reviews['date'].map(lambda x: x[-4:])
```

```python
#subset data so only films from 2010 and beyond are selected
rt_reviews_new = rt_reviews.loc[rt_reviews['year']>='2010']
```
