```python
# Plot of the top 20 movie studios Matplotlib
top_studios.plot(title='Num of Films per Studio', kind='bar')
```

```python
# Seaborn Official Plot of Number of Films Per Studio
ax = sns.barplot(top_studios.index, top_studios.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_title('Number of Films per Studio from 2010 - 2018')
ax.set_xlabel('Studios')
ax.set_ylabel('Number of Films Produced')
plt.tight_layout()
plt.show()
```

```python
bom_movie.head(5)
bom_movie.info()
```

```python
bom_movie_no_nulls = bom_movie.copy()
```

```python
# Drop null values in remaining rows
bom_movie_no_nulls.dropna(axis=0, inplace=True)
```

```python
# Remove the year column, its not relevant to this specific analysis
bom_movie_no_nulls.drop(columns='year', inplace=True)
```

```python
print(bom_movie_no_nulls.info())
bom_movie_no_nulls.head()
```

```python
# Get the average domestic budget for each studio and devide it by a million 
bom_movie_neatened = bom_movie_no_nulls.groupby('studio').mean() /1000000
```

```python
#Rename column to reflect millions fraction
bom_movie_neatened.rename(columns={'domestic_gross': 'dom_gross_in_in_millions'})
print(bom_movie_neatened.head(3))
bom_movie_neatened.info()
```

```python
sorted_gross_studio = bom_movie_neatened.domestic_gross.sort_values(ascending = False)
```

```python
top_gross = sorted_gross_studio.head(20)
```

```python
# Seaborn Official Plot of Top 20 average Grossing Studios by Studio
ax = sns.barplot(top_gross.index, top_gross.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_title('Top Average Gross by Studio 2010 - 2018')
ax.set_xlabel('Studios')
ax.set_ylabel('Gross Domestic in Millions')
#plt.tight_layout()
plt.show()
```

```python
bom_movie_no_nulls.head()
```

```python
top_studios.head(4)
```

```python
# Create df with the gross product and with the number of movies per studio
new_bom_studios = pd.DataFrame([sorted_gross_studio, studio_counts])
#top_studios.join(top_gross, how='left')
new_bom_studios.head()
```

```python
transposed_bom = new_bom_studios.transpose()
transposed_bom.head(5)
```

```python
transposed_bom['ratio'] = transposed_bom.domestic_gross / transposed_bom.studio
transposed_bom.head()
```

```python
transposed_ratio = transposed_bom.ratio.sort_values(ascending = False)
top_ratio = transposed_ratio.head(20)
top_ratio.head()
```

```python
# Seaborn Official Plot of Top 20 ratio of millions made per movie
ax = sns.barplot(top_ratio.index, top_ratio.values)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
ax.set_title('Top Ratio of million gross project per movie by Studio 2010 - 2018')
ax.set_xlabel('Studios')
ax.set_ylabel('Ratio of millions/movie')
plt.tight_layout()
plt.show()
```

```python
imdb_title_crew.rename(columns={'tconst': 'title_id'}, inplace=True)
imdb_title_crew = imdb_title_crew.set_index('title_id')
imdb_title_crew.head(3)
```

```python
imdb_title_principles.rename(columns={'tconst': 'title_id'}, inplace=True)
imdb_title_principles = imdb_title_principles.set_index('title_id')
imdb_title_principles.head(3)
```

```python
imdb_title_ratings.rename(columns={'tconst': 'title_id'}, inplace=True)
imdb_title_ratings = imdb_title_ratings.set_index('title_id')
imdb_title_ratings.head(3)
```

```python
imdb_name_basics.rename(columns={'nconst': 'person_id'}, inplace=True)
imdb_name_basics = imdb_name_basics.set_index('person_id')
imdb_name_basics.head(3)
```

```python
# Now We can join them all together
```

```python
step_1 = imdb_title_principles.join(imdb_title_basics, how='left', on ='title_id')
step_1.head(3)
```

```python
step_2 = step_1.join(imdb_title_crew, how='left', on='title_id')
step_2.head(10)
```

```python
step_3 = step_2.join(imdb_title_ratings, how='left', on='title_id')
step_3.head()
```

```python
step_3.rename(columns={'nconst': 'person_id'})
```

```python
# Now we have a big data set to work with!

final = step_3.join(imdb_name_basics, how='left', on='nconst')
final
```
